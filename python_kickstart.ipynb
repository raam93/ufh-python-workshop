# Building Block 3: Logic & Control Flow

Now we tell Python *when* to do something and *how many times*.

````python
# -- if / elif / else -----------------------------------------------------

filename = "invoice_march.pdf"

if filename.startswith("invoice"):
    print("This is an invoice")
elif filename.startswith("report"):
    print("This is a report")
else:
    print("Unknown type")

# One-line shortcut (ternary expression)
label = "Invoice" if filename.startswith("invoice") else "Other"
print(label)
````

````python
# -- for loop -------------------------------------------------------------

files = ["report.pdf", "budget.xlsx", "photo.jpg", "notes.txt"]

for f in files:
    print(f)

print()

# Loop with index using enumerate()
for i, f in enumerate(files, start=1):
    print(f"{i}. {f}")

print()

# Loop over a range of numbers
for n in range(1, 6):
    print(n, end="  ")
````

````python
# -- while loop -----------------------------------------------------------

attempts = 0
MAX_ATTEMPTS = 3

while attempts < MAX_ATTEMPTS:
    print(f"Attempt {attempts + 1}...")
    attempts += 1

print("Done")
````

````python
# -- try / except ---------------------------------------------------------
# Prevent one bad file from crashing the whole program.

filenames = ["report.pdf", "MISSING.pdf", "budget.xlsx"]

for name in filenames:
    try:
        if "MISSING" in name:
            raise FileNotFoundError("file does not exist")
        print(f"  {name}: OK")
    except FileNotFoundError as e:
        print(f"  {name}: SKIPPED - {e}")
    except Exception as e:
        print(f"  {name}: Unexpected error - {e}")
````

````python
# -- Filter by PREFIX, SUFFIX, or KEYWORD ---------------------------------
import os

folder  = "my_downloads"
PREFIX  = "invoice"   # try: "report", "photo", "backup"
SUFFIX  = ".pdf"      # try: ".xlsx", ".jpg", ".zip"
KEYWORD = "2024"      # try: "q1", "jan", "draft"

print(f'Files starting with "{PREFIX}":')
for filename in os.listdir(folder):
    if filename.lower().startswith(PREFIX):
        print("  -", filename)

print(f'\nFiles ending with "{SUFFIX}":')
for filename in os.listdir(folder):
    if filename.lower().endswith(SUFFIX):
        print("  -", filename)

print(f'\nFiles containing "{KEYWORD}":')
for filename in os.listdir(folder):
    if KEYWORD in filename.lower():
        print("  -", filename)
````

````python
# -- Organize ALL files by type -------------------------------------------
import os, shutil

SOURCE      = "my_downloads"
DESTINATION = "my_downloads_sorted"

FOLDER_RULES = {
    ".pdf":  "PDFs",
    ".jpg":  "Images",  ".jpeg": "Images",  ".png": "Images",
    ".xlsx": "Spreadsheets", ".xls": "Spreadsheets",
    ".docx": "Documents",   ".txt": "Documents",
    ".zip":  "Archives",
}

counts = {}
for filename in os.listdir(SOURCE):
    if os.path.isdir(os.path.join(SOURCE, filename)):
        continue
    ext         = "." + filename.rsplit(".", 1)[-1].lower()
    folder_name = FOLDER_RULES.get(ext, "Other")
    target      = os.path.join(DESTINATION, folder_name)
    os.makedirs(target, exist_ok=True)
    shutil.copy(os.path.join(SOURCE, filename), os.path.join(target, filename))
    counts[folder_name] = counts.get(folder_name, 0) + 1

print("Sorted! Summary:")
for name, n in sorted(counts.items()):
    print(f"  {name:<15} {n} file(s)")
````

# Building Block 4: Functions

A **function** is a reusable, named block of code.

```python
def function_name(param1, param2='default'):
    result = param1 + param2
    return result
```

- `def` starts the definition
- Parameters (also called as "arguments") receive input values when you call the function
- `return` sends a value back to the caller
- Default parameters are used when no value is provided

**Why functions?** Change the logic once, not in five places. Test one small piece at a time.

````python
def add_prefix(filename, prefix):
    return prefix + filename

print(add_prefix("report.pdf", "FINAL_"))
print(add_prefix("IMG_3801.jpg", "vacation_"))


def sequential_name(filename, number, prefix="file_", digits=3):
    ext    = filename.rsplit(".", 1)[-1]
    padded = str(number).zfill(digits)   # "1" -> "001"
    return f"{prefix}{padded}.{ext}"

print(sequential_name("IMG_3801.jpg", 1))
print(sequential_name("IMG_3801.jpg", 1,  prefix="vacation_"))
print(sequential_name("IMG_3801.jpg", 42, prefix="beach_", digits=4))
````

# Task 2: PDF Toolkit

**The problem:** You need to merge three PDFs. You find an online tool, upload your files (now on a stranger's server), download, hope. Or you pay for Adobe.

**The Python solution:** A few lines. Files never leave your machine.

Real cases: merge monthly reports into an archive, extract pages 3–5 to email a colleague, split a scanned form into one file per person.

---
### Running locally
> `pip install pypdf` then point the paths at your real PDF files.

````python
from pypdf import PdfWriter, PdfReader

def merge_pdfs(input_files, output_file="merged.pdf"):
    writer = PdfWriter()
    for path in input_files:
        writer.append(path)
        print(f"  Added: {path}")
    with open(output_file, "wb") as f:
        writer.write(f)
    print(f"Saved -> {output_file}")


def extract_pages(input_file, start_page, end_page, output_file="extracted.pdf"):
    reader = PdfReader(input_file)
    total  = len(reader.pages)
    print(f"{input_file} has {total} pages. Extracting {start_page}-{end_page}.")
    writer = PdfWriter()
    for i in range(start_page - 1, end_page):
        writer.add_page(reader.pages[i])
    with open(output_file, "wb") as f:
        writer.write(f)
    print(f"Saved -> {output_file}")


def split_pdf(input_file, output_folder="split_pages"):
    import os
    os.makedirs(output_folder, exist_ok=True)
    reader = PdfReader(input_file)
    for i, page in enumerate(reader.pages, start=1):
        writer = PdfWriter()
        writer.add_page(page)
        out = os.path.join(output_folder, f"page_{str(i).zfill(3)}.pdf")
        with open(out, "wb") as f:
            writer.write(f)
    print(f"Split {len(reader.pages)} pages -> {output_folder}/")


def merge_all_pdfs_in_folder(folder, output_file="all_combined.pdf"):
    import os
    pdfs  = sorted([f for f in os.listdir(folder) if f.lower().endswith(".pdf")])
    if not pdfs:
        print("No PDFs found in", folder)
        return
    merge_pdfs([os.path.join(folder, f) for f in pdfs], output_file)


````

````python
merge_pdfs(["ufh_laundry_facilities.pdf", "ufh_parking_facilities.pdf"], output_file="all_ufh_facilities.pdf")
extract_pages("all_ufh_facilities.pdf", start_page=1, end_page=3)
split_pdf("ufh_laundry_facilities.pdf", output_folder="facilities_pagewise")
merge_all_pdfs_in_folder("facilities_pagewise", output_file="ufh_facilities_rebuilt.pdf")
````

## Task 3.2: Clean Messy Data

Real data almost always has problems. We fix the five most common:

1. Inconsistent capitalisation (`NORTH` vs `north` vs `North`)
2. Extra whitespace (`East ` with a trailing space)
3. Mixed date formats (`2024-01-15` vs `15/01/2024` vs `Jan 15 2024`)
4. Missing numeric values
5. Duplicate rows

````python
import pandas as pd
df = pd.read_csv("sales_messy.csv")

print("BEFORE:")
print("  Unique regions: ", sorted(df["region"].dropna().unique()))
print("  Missing units:  ", df["units"].isnull().sum())
print("  Duplicate rows: ", df.duplicated().sum())
````

````python
# Fix 1 & 2: Standardise text - strip whitespace, then re-capitalise
# .str applies string methods to an entire column at once.

df["region"]  = df["region"].str.strip().str.lower().str.capitalize()
df["product"] = df["product"].str.strip().str.lower().str.title()
df["notes"]   = df["notes"].str.strip()

print("Regions after:", sorted(df["region"].dropna().unique()))
print("Products after:", sorted(df["product"].dropna().unique()))
````

````python
# Fix 3: Parse mixed date formats into one standard format

df["date"] = pd.to_datetime(df["date"], dayfirst=False, errors="coerce")

print("Date dtype:", df["date"].dtype)
print("Sample:    ", df["date"].head(6).tolist())
print("Unparseable (NaT):", df["date"].isnull().sum())
````

````python
# Fix 4: Fill missing numeric values with the column median
# Median is safer than mean when there may be outliers.

for col in ["units", "unit_price"]:
    n_missing = df[col].isnull().sum()
    df[col] = df[col].fillna(df[col].median())
    print(f"  Filled {n_missing} missing values in {col}")
````

````python
# Fix 5: Remove duplicates and compute the revenue column

before = len(df)
df = df.drop_duplicates()
print(f"Removed {before - len(df)} duplicate rows ({len(df)} remain)")

df["revenue"] = (df["units"] * df["unit_price"]).round(2)

df.to_csv("sales_clean.csv",    index=False)
df.to_excel("sales_clean.xlsx", index=False)
print(f"Saved sales_clean.csv ({len(df)} rows)")
````

## Task 3.4: Combine Multiple Excel Files

Real case: 12 monthly Excel files. You need one master dataset. Python loops through the folder and stacks them — no copy-pasting.

````python
import os
import pandas as pd

def combine_excel_files(folder, output_file="combined.xlsx"):
    excel_files = sorted([f for f in os.listdir(folder) if f.endswith(".xlsx")])
    if not excel_files:
        print("No Excel files found in", folder)
        return None
    frames = []
    for filename in excel_files:
        d = pd.read_excel(os.path.join(folder, filename))
        d["source_file"] = filename
        frames.append(d)
        print(f"  Loaded {filename}: {len(d)} rows")
    combined = pd.concat(frames, ignore_index=True)
    combined.to_excel(output_file, index=False)
    print(f"\nCombined {len(excel_files)} files -> {output_file} ({len(combined)} rows)")
    return combined

df_all = combine_excel_files("monthly_sales", output_file="sales_all_months.xlsx")
````

## Task 3.5: Summary Report

````python
import pandas as pd
df = pd.read_excel("sales_all_months.xlsx")

by_region = df.groupby("region")["revenue"].sum().sort_values(ascending=False)
print("Revenue by region:")
print(by_region.to_string())
````

````python
summary = df.groupby("product").agg(
    total_revenue = ("revenue", "sum"),
    avg_order     = ("revenue", "mean"),
    total_units   = ("units",   "sum"),
    order_count   = ("revenue", "count"),
).round(2).sort_values("total_revenue", ascending=False)
print(summary.to_string())
````

````python
def summary_report(df, group_by, value_col="revenue"):
    report = df.groupby(group_by)[value_col].agg(
        total   = "sum",
        average = "mean",
        count   = "count",
        maximum = "max",
    ).round(2).sort_values("total", ascending=False)
    print(f"\n{value_col.upper()} by {group_by.upper()}")
    print("=" * 55)
    print(report.to_string())
    print(f"Grand total: {df[value_col].sum():,.2f}")
    return report

df = pd.read_excel("sales_all_months.xlsx")
summary_report(df, "region")
summary_report(df, "sales_rep")
summary_report(df, "product", value_col="units")
````

## Task 3.7: Track Changes Between Two Files

Real case: you received a customer list last week and an updated one today. What rows were added or removed? Which values changed?

Doing this manually in Excel is tedious and error-prone. The `DataComparator` class automates it.

````python
import pandas as pd

class DataComparator:
    # Compares two versions of a dataset and reports what changed.
    # Requires an id column that uniquely identifies each row.

    def __init__(self, old_file, new_file, id_col="id"):
        self.old_file = old_file
        self.new_file = new_file
        self.id_col   = id_col
        self.old      = None
        self.new      = None

    def load(self):
        self.old = pd.read_excel(self.old_file)
        self.new = pd.read_excel(self.new_file)
        print(f"Old: {self.old_file} ({len(self.old)} rows)")
        print(f"New: {self.new_file} ({len(self.new)} rows)")
        return self

    def added_rows(self):
        old_ids = set(self.old[self.id_col])
        new_ids = set(self.new[self.id_col])
        return self.new[self.new[self.id_col].isin(new_ids - old_ids)]

    def removed_rows(self):
        old_ids = set(self.old[self.id_col])
        new_ids = set(self.new[self.id_col])
        return self.old[self.old[self.id_col].isin(old_ids - new_ids)]

    def changed_values(self):
        common  = set(self.old[self.id_col]) & set(self.new[self.id_col])
        old_c   = self.old[self.old[self.id_col].isin(common)].set_index(self.id_col)
        new_c   = self.new[self.new[self.id_col].isin(common)].set_index(self.id_col)
        changes = []
        for col in old_c.columns:
            diff = old_c[col] != new_c[col]
            for row_id in diff[diff].index:
                changes.append({
                    self.id_col: row_id,
                    "column":    col,
                    "old_value": old_c.loc[row_id, col],
                    "new_value": new_c.loc[row_id, col],
                })
        return pd.DataFrame(changes)

    def report(self):
        added   = self.added_rows()
        removed = self.removed_rows()
        changed = self.changed_values()
        print("\n" + "=" * 55)
        print("CHANGE REPORT")
        print("=" * 55)
        print(f"\nADDED ({len(added)} rows):")
        print(added.to_string(index=False) if len(added) else "  None")
        print(f"\nREMOVED ({len(removed)} rows):")
        print(removed.to_string(index=False) if len(removed) else "  None")
        print(f"\nCHANGED VALUES ({len(changed)} cells):")
        print(changed.to_string(index=False) if len(changed) else "  None")
        return self


comp = DataComparator("customers_last_week.xlsx", "customers_this_week.xlsx")
comp.load().report()
````

## Task 3.8: DataAnalyst — A Full Pipeline Class

Now we combine cleaning, summarising, and charting into one class. Dirty data goes in; a clean file, a summary, and charts come out.

Each method returns `self`, enabling **method chaining**:
`analyst.load().clean().summarise().plot().save()`

````python
import os
import pandas as pd
import matplotlib.pyplot as plt

class DataAnalyst:
    # Full analysis pipeline: load -> clean -> summarise -> plot -> save.

    def __init__(self, filepath):
        self.filepath = filepath
        self.df       = None
        self.log      = []

    def _note(self, message):
        self.log.append(message)
        print(f"  [{message}]")

    def load(self):
        if self.filepath.endswith(".csv"):
            self.df = pd.read_csv(self.filepath)
        else:
            self.df = pd.read_excel(self.filepath)
        self._note(f"Loaded {self.filepath}: {self.df.shape[0]} rows x {self.df.shape[1]} cols")
        return self

    def clean(self):
        df = self.df
        for col in ["region", "product", "sales_rep"]:
            if col in df.columns:
                df[col] = df[col].str.strip().str.lower().str.capitalize()
        if "date" in df.columns:
            df["date"] = pd.to_datetime(df["date"], errors="coerce")
        for col in df.select_dtypes(include="number").columns:
            missing = df[col].isnull().sum()
            if missing:
                df[col] = df[col].fillna(df[col].median())
                self._note(f"Filled {missing} missing in {col} with median")
        if "units" in df.columns and "unit_price" in df.columns:
            df["revenue"] = (df["units"] * df["unit_price"]).round(2)
        before = len(df)
        df     = df.drop_duplicates()
        if before - len(df):
            self._note(f"Removed {before - len(df)} duplicates")
        self.df = df
        self._note(f"Clean done: {len(self.df)} rows remain")
        return self

    def summarise(self, group_by="region", value="revenue"):
        report = self.df.groupby(group_by)[value].agg(
            total="sum", average="mean", count="count"
        ).round(2).sort_values("total", ascending=False)
        print(f"\nSummary - {value} by {group_by}:")
        print(report.to_string())
        print(f"Grand total: {self.df[value].sum():,.2f}")
        self._note(f"Summarised {value} by {group_by}")
        return self

    def plot(self, group_by="region", value="revenue", save_prefix="output"):
        agg = self.df.groupby(group_by)[value].sum().sort_values(ascending=False)
        plt.figure(figsize=(8, 4))
        agg.plot(kind="bar", color="steelblue", edgecolor="white")
        plt.title(f"{value.capitalize()} by {group_by.capitalize()}")
        plt.ylabel(value.capitalize())
        plt.xticks(rotation=30, ha="right")
        plt.tight_layout()
        out = f"{save_prefix}_{group_by}.png"
        plt.savefig(out, dpi=150)
        plt.show()
        self._note(f"Saved chart: {out}")
        return self

    def save(self, output_file="cleaned_data.xlsx"):
        self.df.to_excel(output_file, index=False)
        self._note(f"Saved to {output_file}")
        print("\n--- Pipeline log ---")
        for entry in self.log:
            print(f"  {entry}")
        return self

print("DataAnalyst class defined.")
````

````python
analyst = DataAnalyst("sales_messy.csv")

(
    analyst
    .load()
    .clean()
    .summarise(group_by="region",  value="revenue")
    .summarise(group_by="product", value="units")
    .plot(group_by="region",  value="revenue", save_prefix="output")
    .plot(group_by="product", value="units",   save_prefix="output")
    .save("sales_final.xlsx")
)
````
